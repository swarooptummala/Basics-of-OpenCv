# Line

###Line

```python
line=cv2.line(image_blank,(100,100),(500,500),(255,0,0),5)
plt.imshow(line)
plt.show()
```

# Rectangle

```python
image_blank = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
rectangle=cv2.rectangle(image_blank,(0,0),(510,510),(0,255,0),10)
plt.imshow(rectangle)
plt.show()
```

# Circle

```python
image_blank = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
circle=cv2.circle(image_blank,(500,500),500,(0,0,255),5)
plt.imshow(circle)
plt.show()
```

# Text

```python
image_blank = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
text = cv2.putText(image_blank,'Swaroop',(10,450),font,3,(0,255,0),20)
plt.imshow(text)
plt.show()
```
